```python
veges = pd.read_csv(r'C:\Users\Rishabh\Desktop\vegetable.csv')
```

```python
l = ['farmer_name']
l.extend(veges.columns[1:])
veges.columns = l
```

```python
veges.head()
```

```python
cust_quan = 0.018
cust_dist = 'sitamarhi'
cust_state = 'bihar'
cust_veg = 'potato'

best_farmer_name = 0

df = veges[veges.dist == cust_dist][veges.state == cust_state]
```

```python
if(cust_quan > sum(df.p_p.values)):
    print("can't deliver so much")
else:
    f_name = []
    if(cust_veg == 'potato'):
        v = "p_p"
    else:
        v = "p_b"
    df = df.sort_values(by = v)
    for i in range(df.shape[0]):
        if(cust_quan > df.p_p.iloc[i]):
            cust_quan -= df.p_p.iloc[i]
            df.p_p.iloc[i] = 0
            f_name.append(df.farmer_name.iloc[i])
        elif(cust_quan != 0):
            df.p_p.iloc[i] -= cust_quan
            cust_quan = 0
            f_name.append(df.farmer_name.iloc[i])
f_name
```
